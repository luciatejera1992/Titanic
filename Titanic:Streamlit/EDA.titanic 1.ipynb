```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

```

```python
df = pd.read_csv(r"titanic.csv")
df.head()
```

```python
df['pclass'].value_counts()
```

```python
df['class'].value_counts()
```

```python
#Calculamos el numero de nulos de la columna embarked
df['embarked'].isnull().sum()
```

```python
#Rellenamos con la moda porque hab√≠a solo dos nulos
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
```

```python
df['embarked'].isnull().sum()
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
df['embark_town'].isnull().sum()
```

```python
df1 = df.copy()
df1.head()
```

```python

df2 = pd.read_csv(r"Titanic-Dataset.csv")
df2.head()
```

```python
df1.columns, df2.columns
```

```python

df1_normalized = df1.rename(columns={

    'survived': 'Survived',

    'pclass': 'Pclass',

    'sex': 'Sex',

    'age': 'Age',

    'sibsp': 'SibSp',

    'parch': 'Parch',

    'fare': 'Fare',

    'embarked': 'Embarked'

})

df1_normalized.head()
```

```python
df1_normalized['PassengerId'] = df2['PassengerId'].values
```

```python
df1_normalized.head()
```

```python


df_combined = pd.merge(df2, df1_normalized, on='PassengerId', how='inner', suffixes=('', '_drop'))

df_combined = df_combined[[col for col in df_combined.columns if not col.endswith('_drop')]]

```

```python
df_combined
```

```python

import re


def get_title(name): 
    title_search = re.search(' ([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)
    return 

#applying get_TITLE function to extract TITLE from name

df_combined['TITLE'] = df_combined['Name'].apply(get_title) 

#printing unique values of TITLE column

df_combined['TITLE'].unique()   
```

```python
#replacing master whit young 

df_combined['TITLE'] = df_combined['TITLE'].replace(['Master'],'Young') 
#replacing 'Ms' and 'Mlle' with 'Miss'
df_combined['TITLE'] = df_combined['TITLE'].replace(['Ms','Mlle'],'Miss') 
#replacing 'Mme' with 'Mrs'
df_combined['TITLE'] = df_combined['TITLE'].replace('Mme','Mrs') 
#replacing 'Don' with 'Royal'
df_combined['TITLE'] = df_combined['TITLE'].replace(['Don','Dona','Lady','Sir','Countess','Jonkheer'],'Royal') 
#replacing 'Major' and 'Col' with 'Military'
df_combined['TITLE'] = df_combined['TITLE'].replace(['Major','Col','Capt'],'Military')
#replacing 'Rev' with 'Priest'
df_combined['TITLE'] = df_combined['TITLE'].replace(['Rev'],'Priest') 
#replacing 'Dr' with 'Medical'
df_combined['TITLE'] = df_combined['TITLE'].replace(['Dr'],'Medical') 

#printing unique values of TITLE column
df_combined['TITLE'].unique()
```

```python
df_combined.head()
```

```python
df_combined['GrupoFamiliar'] = df_combined['SibSp'] + df_combined['Parch'] 
df_combined['GrupoFamiliar'].value_counts()
```

```python
#Relleno los valores nulos de Age con la media de edad de cada Pclass
df_combined['Age'].isnull().sum()
df_combined.groupby('Pclass')['Age'].mean()
df_combined['Age'].fillna(df_combined.groupby('Pclass')['Age'].transform('mean'), inplace=True)
df_combined['Age'].isnull().sum()
```

```python
df_combined['Pclass'].value_counts()
df_combined.groupby('Pclass')['Sex'].value_counts().unstack().plot(kind='bar', stacked=True, figsize=(8,5))
plt.title('Distribuci√≥n de Sexo por Clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```

```python
df_combined['Sex'].value_counts()
plt.figure(figsize=(8,5))
sns.countplot(data=df_combined, x='Sex', palette='Set2')
plt.title('Distribuci√≥n de Sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.show()
```

```python
df_combined['Embarked'].value_counts()
plt.figure(figsize=(8,5))
sns.countplot(data=df_combined, x='Embarked', palette='Set2')
plt.title('Distribuci√≥n de Puerto de Embarque')
plt.xlabel('Puerto de Embarque')
plt.ylabel('Cantidad')
plt.show()
```

```python
df_combined['Survived'].value_counts()
plt.figure(figsize=(8,5))
sns.countplot(data=df_combined, x='Survived', palette='Set2')
plt.title('Distribuci√≥n de Supervivencia')
plt.xlabel('Supervivi√≥ (1) / No Supervivi√≥ (0)')
plt.ylabel('Cantidad')
plt.show()
```

```python
df_combined.groupby('Pclass')['Embarked'].value_counts().unstack().plot(kind='bar', stacked=True, figsize=(8,5))
plt.title('Distribuci√≥n de Puerto de Embarque por Clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```

```python

# Supervivencia por clase
df_combined.groupby('Pclass')['Survived'].value_counts().unstack().plot(kind='bar', stacked=True, figsize=(8,5))
plt.title('Supervivencia por Clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```
